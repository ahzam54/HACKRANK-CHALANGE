```python
if __name__ == '__main__':
    N = int(input())
    listFull = []
    for i in range(N):
        query = input()
        query = query.split()
        match query[0]:
            case 'insert':
                listFull.insert(int(query[1]),int(query[2]))
            case 'print':
                print(listFull)
            case 'remove':
                listFull.remove(int(query[1]))
            case 'append':
                listFull.append(int(query[1]))
            case 'sort':
                listFull.sort()
            case 'pop':
                listFull.pop()
            case 'reverse':
                listFull.reverse()
```
