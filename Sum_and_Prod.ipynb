```python
import numpy

# Read dimensions N and M
N, M = map(int, input().split())

# Read the array elements
my_array = []
for _ in range(N):
    row = list(map(int, input().split()))
    my_array.append(row)

# Convert list to a NumPy array
my_array = numpy.array(my_array)

# 1. Sum along axis 0 (sums columns)
# If the array is [[1, 2], [3, 4]], axis 0 sum is [1+3, 2+4] = [4, 6]
sum_result = numpy.sum(my_array, axis=0)

# 2. Find the product of that sum
# Product of [4, 6] is 4 * 6 = 24
product_result = numpy.prod(sum_result)

print(product_result)
```
