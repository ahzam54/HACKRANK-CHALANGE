```python
from fractions import Fraction
from functools import reduce

def product(fracs):
    t = reduce(lambda x,y:x*y,fracs)
    return t.numerator, t.denominator

if __name__ == '__main__':
    fracs = []
    for _ in range(int(input())):
        fracs.append(Fraction(*map(int, input().split())))
    result = product(fracs)
    print(*result)
```

# Explanation of the Solution
Import reduce: In Python 3, reduce is not a built-in function
anymore. It has been moved to the functools module. You must
import it: from functools import reduce.

The Lambda Function: reduce takes a function that accepts two arguments.
lambda x, y: x * y: This anonymous function takes the accumulated result so far (x) and the next item in the list (y), multiplies them, and returns the new product.

The Process:


*   If your list is [Fraction(1, 2), Fraction(3, 4), Fraction (10, 6)]:


*   Step 1: x = 1/2, y = 3/4 $\rightarrow$ result 3/8.
*   Step 2: x = 3/8, y = 10/6 $\rightarrow$ result 30/48.



The Fraction class automatically simplifies 30/48 to 5/8. List item
